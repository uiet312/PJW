# 데이터 전처리

### 결측치 처리

```python
data.isnull().sum()
```

```python
data = data.fillna(0)
data.head()
```

```python
data.isnull().sum()
```

### 범주형 데이터 변환

```python
data['주구매코너']
```

```python
# 범주형 데이터 변환
obj = ['주구매코너']
data[obj] = data[obj].apply(lambda x: x.astype('category').cat.codes)
```

```python
data['주구매코너']
```

### 파생변수 생성

```python
pd.DataFrame((data.corr()['아침_구매건수']).sort_values(ascending = False))
```

```python
pd.DataFrame((data.corr()['구매건수']).sort_values(ascending = False))
```

- 구매건수 컬럼이 전반적으로 높은 상관도를 보이고 있음 따라서 이를 기준으로 파생변수를 생성

```python
data['아침구매액'] = data['총구매액']*(data['아침_구매건수'] /data['구매건수'])
data['점심구매액'] = data['총구매액']*(data['점심_구매건수'] /data['구매건수'])
data['저녁구매액'] = data['총구매액']*(data['저녁_구매건수'] /data['구매건수']) 
data['봄구매액']   = data['총구매액']*(data['봄_구매건수'] /data['구매건수'])
data['여름구매액'] = data['총구매액']*(data['여름_구매건수'] /data['구매건수']) 
data['가을구매액'] = data['총구매액']*(data['가을_구매건수'] /data['구매건수'])
data['겨울구매액'] = data['총구매액']*(data['겨울_구매건수'] /data['구매건수'])
```

```python
data.head()
```

### 데이터 분할

```python
from sklearn.model_selection import train_test_split

data = data.drop(['총구매액','구매건수','아침_구매건수','점심_구매건수','저녁_구매건수',
                      '봄_구매건수','여름_구매건수','가을_구매건수','겨울_구매건수'], axis=1)

dfX = data.drop(['gender'], axis=1) # axis=1 은 열
dfy = data['gender']

X_train, X_test, y_train, y_test = train_test_split(dfX, dfy, test_size = 0.25, random_state = 0)

display(X_train.shape, X_test.shape)
```

### 데이터 셈플링

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=0)
X_train_over,y_train_over = smote.fit_sample(X_train,y_train)
print('SMOTE 적용 전  데이터 세트: ', X_train.shape, y_train.shape)
print('SMOTE 적용 후  데이터 세트: ', X_train_over.shape, y_train_over.shape)
print('SMOTE 적용 후 데이터 비율: \n', pd.Series(y_train_over).value_counts())
```

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression()

logreg.fit(X_train,y_train).score(X_test, y_test)
```

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression()

logreg.fit(X_train_over,y_train_over).score(X_test, y_test)
```

- 데이터 셈플링 작업을 하였지만 모델의 성능은 감소된걸 알 수 있다.

### 데이터 스케일링
- 특성별로 데이터의 스케일이 다를때 이를 방지

```python
from sklearn.preprocessing import StandardScaler
# 각 feature의 평균을 0, 분상을 1로 변경

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
```

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression()

logreg.fit(X_train_scaled,y_train).score(X_test, y_test)
```

- 데이터 스케일링 작업도 모델의 성능에 좋은 변화를 주지는 못하였다.

# 모델링

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression()

logreg.fit(X_train,y_train).score(X_test, y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier()

tree.fit(X_train,y_train).score(X_test, y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()

knn.fit(X_train,y_train).score(X_test, y_test)
```

```python
from sklearn.ensemble import VotingClassifier

voting = VotingClassifier(
    estimators = [('logreg', logreg), ('tree', tree), ('knn', knn)],
    voting = 'hard')

voting.fit(X_train,y_train).score(X_test, y_test)
```

```python
from sklearn.ensemble import BaggingClassifier

bagging = BaggingClassifier(base_estimator=KNeighborsClassifier(), random_state=0, n_estimators=200)
bagging.fit(X_train, y_train).score(X_test, y_test)
```

```python
from sklearn.ensemble import AdaBoostClassifier

ada = AdaBoostClassifier(n_estimators=200, random_state=0)
ada.fit(X_train, y_train).score(X_test, y_test)
```

```python
from sklearn.ensemble import GradientBoostingClassifier

gbm = GradientBoostingClassifier(n_estimators=200, random_state=0)
gbm.fit(X_train, y_train).score(X_test, y_test)
```

```python
from lightgbm import LGBMClassifier

lgb = LGBMClassifier(n_estimaotrs = 200, random_state=0)
lgb.fit(X_train, y_train).score(X_test, y_test)
```

# 데이터 검증

```python
from sklearn.model_selection import cross_val_score
from sklearn.svm import SVC

model = SVC()
scores = cross_val_score(model, X_train, y_train); scores
```

```python
print("Mean: {:.3f}\nStd: {:.3f}\nMin: {:.3f}\nMax: {:.3f}".format(
    scores.mean(), scores.std(), scores.min(), scores.max()))
# 분산이 작으면 Overfitting 이 적다
# 과적합 : 100~90   적정적합 : 90~65    과소적합 : 65~
```
